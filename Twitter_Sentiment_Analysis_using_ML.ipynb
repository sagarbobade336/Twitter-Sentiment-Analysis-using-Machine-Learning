Data Processing

```python
twitter_data=pd.read_csv(f'{path}/training.1600000.processed.noemoticon.csv', encoding='latin1')
```

```python
#Naming the column
twitter_data.columns=['target', 'ids', 'date', 'flag', 'user', 'text']
```

Convert the target "4" to "1"

```python
twitter_data.replace({"target":{4:1}}, inplace=True)
```

```python
twitter_data["target"].value_counts()
```

0->Negative Tweet

1->Positive Tweet

**stemming** is process of reducing a word to its Root word

example:actor,aactress,acting=act

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):

  stemmed_content = re.sub('[^a-zA-Z]','.',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
twitter_data["stemmed_content"]=twitter_data["text"].apply(stemming)
```

```python
twitter_data.head()
```

```python
twitter_data["target"]
```

```python
X = twitter_data["stemmed_content"].values
Y = twitter_data["target"].values
```

```python
X
```

```python
Y
```

Spliting the data to traning data and the test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X.shape, Y.shape, X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
#
```

```python
X_train
```

```python
X_test
```

```python
# converting the textual data to numerical data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
X_train
```

```python
X_test
```

Training the Machine Learining Model

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evalution

Accuarcy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data : ", training_data_accuracy)
```

Saving the trained model

```python
import pickle
```

```python
filename = "trained_model.sav"
pickle.dump(model,open(filename,'wb'))
```
